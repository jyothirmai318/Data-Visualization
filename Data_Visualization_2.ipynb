For this, I have used Automobile Dataset for data analysis. This Dataset has different characteristics such as body-style, wheel-base, engine-type, price, mileage, horsepower, etc.

```python
#cleaning the dataset and replacing null values
import pandas as pd
df = pd.read_csv('Automobile_data.csv')
df.replace(["?", "n.a", "NaN"], None, inplace=True)
cleaned_file_name = 'cleaned_Automobile_data.csv'
df.to_csv(cleaned_file_name, index=False)
```

```python
#Highest priced car
df = pd.read_csv('cleaned_Automobile_data.csv')
company_highest_price = df.groupby('company')['price'].max()
print(company_highest_price)
```

```python
#average mileage of each car making company.
average_mileage = df.groupby('company')['average-mileage'].mean()
print(average_mileage)
```

#####Concatenating two data frames using the following conditions. Create two data frames using the following two dictionaries.

#####japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}

```python
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925, 71400]}
JapaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi'], 'Price': [29995, 23600, 61500, 58900]}
df_german = pd.DataFrame(GermanCars)
df_japanese = pd.DataFrame(JapaneseCars)
df1= pd.concat([df_german, df_japanese], ignore_index=True)
print(df1)
```

#####Merging two data frames using the following condition. Create two data frames using the following two Dicts, Merge two data frames, and append the second data frame as a new column to the first data frame.


#####Get total profit of all months and show line plot with the following Style properties
Generated a line plot with the following Style properties: â€“
*  Line Style:dotted and Line-color: red
*  Legend at the lower right location.
*  X label name = Month Number
*  Y label name = Sold units number
*  A circle marker.

```python
import matplotlib.pyplot as plt
df = pd.read_csv('company_sales_data.csv')
plt.figure(figsize=(10, 6))
plt.plot(df['month_number'], df['total_profit'], linestyle=':', color='red',marker='o',markerfacecolor='red',linewidth=3)
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.title('Total Profit vs. Month Number')
plt.legend(['Total Profit'], loc='lower right')
plt.show()
```

#####Making a plot of the function f(x)=x^2andg(x)=x^3forâˆ’5â‰¤xâ‰¤5.

```python
import seaborn as sns
import numpy as np

sns.set()
x = np.linspace(-5, 5, 100)
f_x = x**2
g_x = x**3
plt.figure(figsize=(10, 6))
plt.plot(x, f_x, color='red', marker='o', label='$f(x) = x^2$')
plt.plot(x, g_x, color='blue', marker='s', label='$g(x) = x^3$')
plt.title('Plot of $f(x) = x^2$ and $g(x) = x^3$', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.legend(loc=10)
plt.xlim(-6, 6)
plt.ylim(-10, 10)
plt.grid(True)
plt.show()
```

#####Given the lists x = np.arange(11) and y=x^2, creating a 2 by 3 subplot where each subplot plots x versus y using plot, scatter, bar, loglog, semilogx, and semilogy.

```python
x = np.arange(11)
y = x**2
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
axs[0, 0].plot(x, y, label="x vs x^2", color='blue')
axs[0, 0].set_title('Plot: x vs x^2')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')
axs[0, 0].grid(True)

axs[0, 1].scatter(x, y, color='red')
axs[0, 1].set_title('Scatter: x vs x^2')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')
axs[0, 1].grid(True)

axs[0, 2].bar(x, y, color='green')
axs[0, 2].set_title('Bar: x vs x^2')
axs[0, 2].set_xlabel('x')
axs[0, 2].set_ylabel('y')
axs[0, 2].grid(True)

axs[1, 0].loglog(x, y, color='purple')
axs[1, 0].set_title('Log-Log: x vs x^2')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
axs[1, 0].grid(True)

axs[1, 1].semilogx(x, y, color='orange')
axs[1, 1].set_title('Semilogx: x vs x^2')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')
axs[1, 1].grid(True)

axs[1, 2].semilogy(x, y, color='brown')
axs[1, 2].set_title('Semilogy: x vs x^2')
axs[1, 2].set_xlabel('x')
axs[1, 2].set_ylabel('y')
axs[1, 2].grid(True)

plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

zdata_group_1 = 15 * np.random.random(100)
xdata_group_1 = np.sin(zdata_group_1) + 0.1 * np.random.randn(100)
ydata_group_1 = np.cos(zdata_group_1) + 0.1 * np.random.randn(100)

zdata_group_2 = [10, 3, 5, 8, 9, 6]
xdata_group_2 = [4, 9, 9, 5, 10, 8]
ydata_group_2 = [5, 5, 2, 10, 5, 3]

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xdata_group_1, ydata_group_1, zdata_group_1, color='green', marker='o', label='Group 1')
ax.scatter(xdata_group_2, ydata_group_2, zdata_group_2, color='red', marker='+', label='Group 2')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Scatter Plot of Two Groups')
ax.legend()
plt.show()
```

#####The function to plot the 3D surfaces is plot_surface(X,Y,Z), where X and Y are the output arrays from meshgrid, and Z=f(X,Y) or Z(i,j)=f(X(i,j),Y(i,j)). The most common surface plotting functions are surf and contour.
#####Plotting of the surface f(x,y)=sin(x)â‹…cos(y)forâˆ’5â‰¤xâ‰¤5,âˆ’5â‰¤yâ‰¤5 using the plot_surface function.

```python

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

fig = plt.figure(figsize=(12,6))

ax1 = fig.add_subplot(121, projection='3d')
contour = ax1.contour(X, Y, Z, cmap='viridis')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')

ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X, Y, Z, cmap='viridis')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')

plt.tight_layout()
plt.show()
```
